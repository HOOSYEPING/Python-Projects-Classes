```python
#Practice Exercise: Extracting State Population
import requests
api_url="https://jian.sh/malaysia-api/state/v1/all.json"
response = requests.get(api_url)
malaysia_state_data=response.json()

user_input=input("Enter the name of a state :")

for state in malaysia_state_data:
    if state['state'].lower()==user_input.lower():
        print (f"Population of {user_input} :{state['population']}") #to print answer that related to previous input data use{}
        break
        
else:
    print (f"State '{user_input}' not found in the data.")
```

```python
#Filter Data: Filter out studnets who have a grade less than 'B'.
import csv 

input_csv_file = 'copy of students.csv'
output_csv_file = 'top_students.csv'
filtered_criteria = ['A', 'B']

filtered_rows = []

with open(input_csv_file, 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        if row['Grade'] in filtered_criteria:
            row['Status']='Passed'
            filtered_rows.append([row['Name'],row['Grade'],row['Status']])
           
with open(output_csv_file, 'w', newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['Name','Grade','Status'])
    writer.writerows(filtered_rows)
```
